# How to extract text from .doc / .docx

## Using XML 

```python
#specific to extracting information from word documents
import os
print(os.getcwd())
import zipfile
#other tools useful in extracting the information from our document
import re
#to pretty print our xml:
import xml.dom.minidom

docx_filename = './samples/docx/Alliance Management.docx'
doc_filename= './samples/doc/20190507-USFJ-J00-Guidance.doc'


document = zipfile.ZipFile(docx_filename)
# print(document)
print(document.namelist())
name = 'word/document.xml'
uglyXml = xml.dom.minidom.parseString(document.read(name)).toprettyxml(indent='  ')

text_re = re.compile('>\n\s+([^<>\s].*?)\n\s+</', re.DOTALL)    
prettyXml = text_re.sub('>\g<1></', uglyXml)

print(prettyXml.lstrip())

```

## Direct to .txt

```python
import docx2txt

# Passing docx file to process function
text = docx2txt.process(docx_filename)
#print(text)
out_file = doc_filename.rstrip('docx')+'txt'
# Saving content inside docx file into output.txt file
with open(out_file, "w") as text_file:
	print(text, file=text_file)
```

## .doc to .txt

```python

from glob import glob
import re
import os
import win32com.client as win32
from win32com.client import constants
import win32com

# Create list of paths to .doc files
path = glob(doc_filename, recursive=True)

def save_as_docx(path):
    # Opening MS Word
    word = win32.gencache.EnsureDispatch('Word.Application')
    doc = word.Documents.OpenOld(path)
    doc.Activate ()

    # Rename path with .docx
    new_file_abs = os.path.abspath(path)
    new_file_abs = re.sub(r'\.\w+$', '.docx', new_file_abs)

    # Save and Close
    word.ActiveDocument.SaveAs(
        new_file_abs, FileFormat=constants.wdFormatXMLDocument
    )
    doc.Close(False)

# for path in paths:
#     save_as_docx(path)
#save_as_docx(path)
```

## pptx to txt

```python
path ='c://Users//ituki//Documents//Projects//NLP//samples//pptx'

from pptx import Presentation
import glob

def pptx2txt(path, PRINT=False):
    curdir = os.curdir
    os.chdir(path)
    #file_names = glob.glob('*pptx*', recursive=True)
    txt_list=[]
    for eachfile in glob.glob("*.pptx"):
        prs = Presentation(eachfile)
        print(eachfile)
        print("----------------------")
        txt = ""
        for slide in prs.slides:
            for shape in slide.shapes:
                if hasattr(shape, "text"):
                    if PRINT:
                        print(shape.text)
                    txt+=shape.text
        if txt == "":
            print('error')
            break
        txt_list.append(txt)
    os.chdir(curdir)
    return txt_list
```
